```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
end = "act"

sing_left = np.genfromtxt(f"data/singlet_left_{end}")
sing_right = np.genfromtxt(f"data/singlet_right_{end}")
trip_left = np.genfromtxt(f"data/triplet_left_{end}")
trip_right = np.genfromtxt(f"data/triplet_right_{end}")

sing = np.append(sing_left[::-1].T, sing_right[1:].T, axis=1)
trip = np.append(trip_left[::-1].T, trip_right[1:].T, axis=1)

energy_combined = np.where(sing[1]-trip[1] < 0, sing[1], trip[1])

```

```python
fig, ax = plt.subplots()
ax.plot(sing[0], sing[1])
ax.plot(trip[0], trip[1])
ax.plot(trip[0], energy_combined)
fig.show();
```

```python
np.savetxt("energy.dat", np.column_stack((trip[0], energy_combined)))
```
